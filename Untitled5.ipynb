```python
# 🔗 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 📁 2. Set Dataset Path (CHANGE if needed)
data_path = "/content/drive/MyDrive/archive.zip"

# -------------------------------
# 📦 Imports
# --------------------------------
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import zipfile

from torchvision import datasets, transforms
from torch.utils.data import random_split, DataLoader

# -------------------------------
# 🔄 Unzip the dataset if it's a zip file and set correct data_path
# -------------------------------
if data_path.endswith('.zip'):
    extract_base_path = '/content/plant_disease_dataset_extracted'
    os.makedirs(extract_base_path, exist_ok=True)
    with zipfile.ZipFile(data_path, 'r') as zip_ref:
        zip_ref.extractall(extract_base_path)

    # Find the actual root of image classes
    # The structure often is archive.zip -> dataset_folder -> color/grayscale -> class_folders
    # We need to navigate to the 'color' folder for image processing

    # List contents of the extracted base path
    first_level_contents = os.listdir(extract_base_path)
    if len(first_level_contents) == 1:
        # Assuming there's one main folder, e.g., 'plantvillage dataset'
        nested_path = os.path.join(extract_base_path, first_level_contents[0])
        if os.path.isdir(nested_path):
            second_level_contents = os.listdir(nested_path)
            if 'color' in second_level_contents:
                data_path = os.path.join(nested_path, 'color')
            elif 'grayscale' in second_level_contents:
                data_path = os.path.join(nested_path, 'grayscale')
            else:
                data_path = nested_path # Fallback if no color/grayscale, assume classes are here
        else:
            data_path = extract_base_path # Fallback if first_level_contents[0] is not a dir
    else:
        data_path = extract_base_path # Fallback if multiple items at first level, assume classes are here

    print(f"Dataset processed, data_path set to: {data_path}")

# -------------------------------
# ✅ 3. Class Names + Count
# -------------------------------
classes = os.listdir(data_path)

class_counts = {}
# Filter out any non-directory entries (like .DS_Store, or other files)
classes = [cls for cls in classes if os.path.isdir(os.path.join(data_path, cls))]

for cls in classes:
    class_counts[cls] = len(os.listdir(os.path.join(data_path, cls)))

df = pd.DataFrame(class_counts.items(), columns=["Class", "Count"])
df = df.sort_values(by="Count", ascending=False)

print("Total Classes:", len(classes))
print(df.head())

# -------------------------------
# 📊 4. Plot Class Distribution
# -------------------------------
plt.figure(figsize=(12,5))
plt.bar(df["Class"], df["Count"])
plt.xticks(rotation=90)
plt.title("Class Distribution")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.show()

# -------------------------------
# 🖼 5. Show Sample Images
# -------------------------------
plt.figure(figsize=(10,6))

for i, cls in enumerate(classes[:min(len(classes), 6)]): # Ensure we don't try to plot more than available classes
    folder = os.path.join(data_path, cls)
    # Ensure the folder contains files, not empty
    if os.path.isdir(folder) and os.listdir(folder):
        img_path = os.path.join(folder, random.choice(os.listdir(folder)))
        try:
            img = Image.open(img_path)
            plt.subplot(2,3,i+1)
            plt.imshow(img)
            plt.title(cls)
            plt.axis("off")
        except Exception as e:
            print(f"Could not open image {img_path}: {e}")
    else:
        print(f"Class folder {folder} is empty or not a directory, skipping sample image.")

plt.tight_layout()
plt.show()

# -------------------------------
# ⚙️ 6. Preprocessing for CNN
# -------------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

dataset = datasets.ImageFolder(data_path, transform=transform)

# -------------------------------
# ✂️ 7. Train / Val / Test Split
# -------------------------------
train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_data, val_data, test_data = random_split(dataset, [train_size, val_size, test_size])

print("Train:", len(train_data))
print("Validation:", len(val_data))
print("Test:", len(test_data))

# -------------------------------
# 🔄 8. DataLoaders
# -------------------------------
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32)
test_loader = DataLoader(test_data, batch_size=32)

print("✅ Data ready for CNN training")
```

```python
# =========================
# 1. MOUNT GOOGLE DRIVE
# =========================
from google.colab import drive
drive.mount('/content/drive')

# =========================
# 2. IMPORT LIBRARIES
# =========================
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# =========================
# 3. CONFIG
# =========================
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 25

# 🔁 CHANGE THIS PATH (example below)
DATASET_PATH = "/content/drive/MyDrive/plant_dataset"

# =========================
# 4. DATA PREPROCESSING
# =========================
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    zoom_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_data = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

NUM_CLASSES = train_data.num_classes
print("Classes:", train_data.class_indices)

# =========================
# 5. MODEL (TRANSFER LEARNING)
# =========================
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights='imagenet'
)

base_model.trainable = False

x = base_model.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.5)(x)

outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)

model = models.Model(inputs=base_model.input, outputs=outputs)

# =========================
# 6. COMPILE
# =========================
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# =========================
# 7. CALLBACKS
# =========================
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(patience=3),
    tf.keras.callbacks.ModelCheckpoint(
        "/content/drive/MyDrive/best_model.h5",
        save_best_only=True
    )
]

# =========================
# 8. TRAIN
# =========================
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS,
    callbacks=callbacks
)

# =========================
# 9. SAVE MODEL TO DRIVE
# =========================
model.save("/content/drive/MyDrive/plant_disease_model.h5")

# =========================
# 10. EVALUATION
# =========================
loss, acc = model.evaluate(val_data)
print(f"Validation Accuracy: {acc*100:.2f}%")
```
